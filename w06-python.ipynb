```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(2023)

N = 80
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (20 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.7)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-1000, 1000, 10000)  # 10,000   
y = np.random.randint(-1000, 1000, 10000)  # 10,000   

mask1 = abs(x) > 50  # x  50   
mask2 = abs(y) > 50  # y  50   

x = x[mask1 + mask2]  # mask1 mask2  x 
y = y[mask1 + mask2]  # mask1 mask2  y 

plt.scatter(x, y, c=x, cmap='jet', alpha=0.4)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-1000, 1000, 10000)
y = np.random.randint(-1000, 1000, 10000)

mask1 = abs(x) > 500
mask2 = abs(y) > 500
x = x[mask1 & mask2]
y = y[mask1 & mask2]

plt.scatter(x, y, c=x, cmap='jet', alpha=0.4)
plt.colorbar()
plt.show()
```
